```python
import tensorflow as tf

# Step 1: Create a random tensor of shape (4, 6)
tensor = tf.random.uniform((4, 6))

# Step 2: Find its rank and shape
rank = tf.rank(tensor)
shape = tf.shape(tensor)

print(f"Original Tensor Rank: {rank.numpy()}")
print(f"Original Tensor Shape: {shape.numpy()}")

# Step 3: Reshape to (2, 3, 4)
reshaped_tensor = tf.reshape(tensor, (2, 3, 4))
reshaped_shape = tf.shape(reshaped_tensor)

# Transpose to (3, 2, 4)
transposed_tensor = tf.transpose(reshaped_tensor, perm=[1, 0, 2])
transposed_shape = tf.shape(transposed_tensor)

print(f"Reshaped Tensor Shape: {reshaped_shape.numpy()}")
print(f"Transposed Tensor Shape: {transposed_shape.numpy()}")

# Step 4: Broadcast a smaller tensor (1, 4) to match the larger tensor and add them
small_tensor = tf.random.uniform((4, 1))
broadcasted_tensor = tf.broadcast_to(small_tensor, (4, 6))

added_tensor = tensor + broadcasted_tensor  # Broadcasting happens automatically

print(f"Broadcasted Tensor Shape: {broadcasted_tensor.shape}")
print(f"Added Tensor Shape: {added_tensor.shape}")

# Step 5: Explanation of broadcasting
"""
Broadcasting in TensorFlow:
- Allows tensors of different shapes to be used in element-wise operations.
- The smaller tensor (4,1) expands its second dimension to match (4,6).
- This process avoids unnecessary memory duplication.
"""
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define true values (y_true) and model predictions (y_pred)
y_true = np.array([0, 1, 0])  # One-hot encoded class label (e.g., class 2)
y_pred1 = np.array([0.1, 0.8, 0.1])  # Predicted probabilities (close to true)
y_pred2 = np.array([0.3, 0.4, 0.3])  # Slightly modified prediction (less confident)

# Convert to TensorFlow tensors
y_true_tf = tf.convert_to_tensor(y_true, dtype=tf.float32)
y_pred1_tf = tf.convert_to_tensor(y_pred1, dtype=tf.float32)
y_pred2_tf = tf.convert_to_tensor(y_pred2, dtype=tf.float32)

# Step 2: Compute MSE Loss
mse = tf.keras.losses.MeanSquaredError()
mse_loss1 = mse(y_true_tf, y_pred1_tf).numpy()
mse_loss2 = mse(y_true_tf, y_pred2_tf).numpy()

# Step 3: Compute Categorical Cross-Entropy (CCE) Loss
cce = tf.keras.losses.CategoricalCrossentropy()
cce_loss1 = cce(y_true_tf, y_pred1_tf).numpy()
cce_loss2 = cce(y_true_tf, y_pred2_tf).numpy()

# Step 4: Print Loss Values
print(f"Loss Values for First Prediction:")
print(f" - MSE: {mse_loss1:.4f}")
print(f" - Categorical Cross-Entropy: {cce_loss1:.4f}\n")

print(f"Loss Values for Modified Prediction:")
print(f" - MSE: {mse_loss2:.4f}")
print(f" - Categorical Cross-Entropy: {cce_loss2:.4f}")

# Step 5: Plot Loss Comparison
loss_labels = ["MSE (y_pred1)", "MSE (y_pred2)", "CCE (y_pred1)", "CCE (y_pred2)"]
loss_values = [mse_loss1, mse_loss2, cce_loss1, cce_loss2]

plt.figure(figsize=(8, 5))
plt.bar(loss_labels, loss_values, color=['blue', 'lightblue', 'red', 'salmon'])
plt.xlabel("Loss Type")
plt.ylabel("Loss Value")
plt.title("Comparison of MSE and Categorical Cross-Entropy Loss")
plt.ylim(0, max(loss_values) + 0.1)
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Step 1: Load the MNIST dataset
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize pixel values to the range [0,1]
x_train, x_test = x_train / 255.0, x_test / 255.0

# Step 2: Build a simple neural network
def create_model():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(input_shape=(28, 28)),  # Flatten 28x28 images
        tf.keras.layers.Dense(128, activation='relu'),  # Hidden layer
        tf.keras.layers.Dropout(0.2),  # Dropout to prevent overfitting
        tf.keras.layers.Dense(10, activation='softmax')  # Output layer (10 classes)
    ])
    return model

# Step 3: Train the model using Adam optimizer
adam_model = create_model()
adam_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
adam_history = adam_model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=32, verbose=1)

# Step 4: Train the model using SGD optimizer
sgd_model = create_model()
sgd_model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
sgd_history = sgd_model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=32, verbose=1)

# Step 5: Compare training and validation accuracy
plt.figure(figsize=(10, 5))
plt.plot(adam_history.history['accuracy'], label="Adam - Train Accuracy", color='blue')
plt.plot(adam_history.history['val_accuracy'], label="Adam - Val Accuracy", linestyle="dashed", color='blue')
plt.plot(sgd_history.history['accuracy'], label="SGD - Train Accuracy", color='red')
plt.plot(sgd_history.history['val_accuracy'], label="SGD - Val Accuracy", linestyle="dashed", color='red')

plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Adam vs. SGD: Training and Validation Accuracy")
plt.legend()
plt.show()
```

```python
import tensorflow as tf
import datetime  # For timestamping logs

# Step 1: Load and preprocess the MNIST dataset
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize the images to [0,1] range
x_train, x_test = x_train / 255.0, x_test / 255.0

# Step 2: Create a simple neural network
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),  # Flatten input images
    tf.keras.layers.Dense(128, activation='relu'),  # Hidden layer
    tf.keras.layers.Dropout(0.2),  # Prevent overfitting
    tf.keras.layers.Dense(10, activation='softmax')  # Output layer (10 classes)
])

# Step 3: Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Step 4: Enable TensorBoard logging
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")  # Unique log directory
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

# Step 5: Train the model with TensorBoard logging
model.fit(x_train, y_train, validation_data=(x_test, y_test),
          epochs=5, batch_size=32, callbacks=[tensorboard_callback], verbose=1)

print(f"Training complete. Logs stored in: {log_dir}")
```

2.	How can you use TensorBoard to detect overfitting?

Here's a simpler breakdown of detecting overfitting using TensorBoard:

Monitor Accuracy
- Use TensorBoard's Scalars tab
- Warning sign: Training accuracy rises while validation stays flat
  Example: Training 99% vs Validation 85%
- Fix: Add dropout layers or stop training earlier

Check Loss Values
- Both losses should decrease together
- Red flag: Training loss drops while validation loss grows
  Example: Training 0.01 vs Validation climbing (0.3 → 0.4 → 0.5)
- Fix: Simplify model or add regularization

Watch Weight Patterns
- Use Histograms tab
- Problem: Extreme or concentrated weights
- Fix: Add L2 regularization to control weight size

Key Steps:
1. Launch TensorBoard (`tensorboard --logdir=logs/fit`)
2. Compare training vs validation metrics
3. Adjust model based on patterns seen

3.What happens when you increase the number of epochs?
Here's a simpler explanation of the three main phases in model training:

1. Learning Phase (Early)
- Both accuracies improve
- Loss decreases
- Example: Training 70% → 85%
- Good sign: Model is learning properly

2. Plateau Phase (Middle)
- Improvements slow down
- Little change in metrics
- Example: Training ~95%, Val ~90%
- Normal: Model has learned most patterns

3. Overfitting Phase (Late)
- Training accuracy keeps rising
- Validation accuracy drops
- Training loss decreases while validation loss grows
- Problem: Model is memorizing, not learning

Think of it like studying for a test:
- First you learn quickly
- Then progress slows as you master basics
- Studying too long might make you memorize exact questions instead of understanding concepts
